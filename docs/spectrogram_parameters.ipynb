# Spectrogram Creator parameters and settings

This document explains the technical details behind the Spectrogram Creator tool.

### The Mathematical Concept: Using the ΔMi−1 Formula

A typical additive synthesizer looks at a column of pixels, generates a pure sine wave for each active pixel, and adds them up. That produces a very cold, predictable sound. 

To recreate an "Aphex Twin"-style organic generation, the engine uses the ΔMi−1 equation as a **coupled particle physics simulator**, translated to sound:

ΔMi = -Di · [Σ(j∈C{i}) Fji(Mj - Mi)] + Fext_i

When you change settings, you are directly manipulating these physics variables.

### Physics Parameters (The ΔMi−1 Variables)

These sliders determine how frequency bins "talk" to each other. They change the texture, resonance, and bleed of the sound—effectively changing the "material" the sound is made of.

**1. Damping (Di)**
*   **What it does:** Determines how quickly a frequency's amplitude decays back to zero after it has been triggered. 
*   **Low (e.g., 0.01 - 0.05):** Frequencies ring out for a long time, creating an echoy, smudged, ambient, or "wet" sound. The transitions between different parts of the drawing will bleed together.
*   **High (e.g., 0.3 - 0.5):** Extremely fast decay. The sound will be much drier, punchier, and distinctly follow the exact shapes you've painted with almost zero resonance trail.

**2. Coupling Strength (Fji)**
*   **What it does:** Dictates how strongly neighboring frequency bins pull on one another. The synthesizer mimics a physical surface (like a drum head or a vocal cord). If one frequency is vibrating very loudly, it "drags" its neighbors along with it.
*   **Low (e.g., 0.00):** Pure additive synthesis. Frequencies are completely independent. A straight line you draw sounds exactly like a clean, artificial test tone.
*   **High (e.g., 0.15 - 0.30):** High physical tension. Strong fundamental tones will excite the frequencies immediately above and below them, introducing natural harmonic distortion, warmth, and "fizz."

**3. Coupling Radius (C{i})**
*   **What it does:** Sets the physical distance of the coupling effect. 
*   **Low (e.g., 1-2):** A frequency only interacts with the bin immediately adjacent to it. You get a very tight, narrow band of distortion.
*   **High (e.g., 8-10):** A frequency interacts with several bins far above and below it. Painting a narrow line will create a "fat", broadband, buzzing, almost vocal-like texture because the energy spreads across a wide chunk of the spectrum.

**4. Force Gain (Fext Gain)**
*   **What it does:** Amplifies or attenuates the external force (your drawing and live audio input) before it goes into the physics equation. It acts as an energy multiplier.
*   **Low (e.g., 0.1x - 0.5x):** Your drawing acts as a gentle whisper into the system. It will barely excite the coupled frequencies unless they are drawn very intensely.
*   **High (e.g., 2.0x - 3.0x):** Your drawing blasts the system with energy. Even faint lines will forcefully resonate, bleeding heavily into neighboring bins and quickly reaching maximum brightness (clipping the local amplitude limits).

**5. Simulation ON/OFF**
*   **What it does:** Toggles whether you see the raw input data (just the pixels you drew and the live track energy) or the **physics-evolved grid** (the final output after applying the ΔMi−1 formula to all time slices).
*   **When ON:** The canvas visually smears, bleeds, and reacts in real-time as you drag any of the physics sliders, letting you "see" the sound's texture before you hit play.
*   **When OFF:** Shows exactly what you've painted and captured in a crisp, 1-to-1 manner without acoustic decay or coupling being visually rendered.

### The Drawing Toolbar (External Force / Fext_i)

The toolbar determines what external energy you are "injecting" into the physics simulator. You are drawing the Fext_i part of the equation.

**1. Drawing Tools:**
*   **Brush:** Paints raw energy into the spectrum. Since the horizontal axis is time and vertical is frequency (Log Scale: 20Hz at the bottom to 16kHz at the top), moving the brush up and down creates sweeping, whistling sounds (glissandos).
*   **Harmonic:** Instead of just painting one pixel, it automatically paints that pixel *and its mathematical overtones* (x*2, x*3, x*4, etc.) vertically above it. This means every stroke sounds like a musical instrument (like a synth pluck) rather than a pure sine wave beep.
*   **Eraser:** Removes energy from the simulation.

**2. Size:**
*   Controls the radius of the brush. A small size (1-2) makes razor-thin sine waves or sharp sweeps. A large size paints "blocks" of sound, resulting in walls of broadband rushing wind or noise. 

**3. Intensity:**
*   **Mapping:** Maps to the color brightness from dark purple (quiet) to bright yellow/white (loud). 
*   **Effect:** It defines how much energy goes into the formula. If intensity is low, the resulting sound is faint and might struggle to keep the "coupled oscillators" moving. If high, it hits the maximum volume cap and maximally excites neighboring bins.

**4. Duration:**
*   Controls the total playback length (1 to 15 seconds) over the horizontal axis. 
*   **Short (1s):** The playhead moves rapidly across the canvas. A shape crossing the whole screen translates to extremely fast, zipper-like sweeps. 
*   **Long (15s):** The playhead creeps along. That same shape drawn across the screen becomes a slow, evolving, drone-like texture.

**5. Live Mode Button & Now Playing Indicator:**
*   **What it does:** Toggles a real-time visual and audio connection to the application's music player. Next to the button, a **Now Playing** indicator blinks green to confirm which song's spectrum is actively being captured.
*   **Visual Effect:** A vertical red write-head scrolls across the canvas, permanently burning the active frequency spectrum of the currently playing song right into the grid.
*   **How to use it:** You can use this to sample or remix existing songs. If you see the heavy bass lines or snare hits of a playing song mapped on the screen, you can draw using the `Brush` or `Harmonic` tools directly *over* those live patterns to add your own synthesized layers. Because the **physics simulation** is active on the canvas, dragging sliders (like damping and coupling) literally warps the visual footprint and resulting audio resonance of the recorded song in real time!

### Remixing Songs in Real-Time

With Live Mode, play any song in the app and the Spectrogram Creator captures its raw frequency energy onto the canvas. You can then:

- Paint new tones on top of the song's spectrum to add harmonics

- Use the Harmonic brush to layer overtones that blend with the track

- Drag physics sliders to warp how the captured energy resonates — turning a pop song into an ambient drone or a glitchy texture
